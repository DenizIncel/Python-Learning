```python
import pandas as pd

link = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

titanic_df = pd.read_csv(link)
```

Pandas'ta filtreleme yaparken tabloya adeta bir soru sorarız ve o sorunun cevabı olan satırları çekip alırız. Örneğin, sadece 1. sınıf bilete sahip yolcuları (Pclass sütunu) filtrelemek istesek kalıbımız şuna benzer:

```python
# "Önce kadınlar ve çocuklar" demek için kadınları ve çocukları seçelim.

cocuklar = titanic_df["Age"] < 18
kadinlar = titanic_df["Sex"] == "female"

print(titanic_df[cocuklar | kadinlar].head())
```

**iloc (Integer Location - Sayısal Konum)**: Sadece sıra numaralarıyla çalışır. Klasik Python listelerindeki [0] mantığıdır. İsimlerle asla ilgilenmez.

**loc (Location - İsim/Etiket Konumu)**: Satırın indeksi veya sütunun tam adıyla çalışır.

```python
ilk_yolcunun_bilgisi = titanic_df.iloc[0 , 0:4]  # İlk yolcunun ilk 4 sütunundaki bilgileri aldık.
print(ilk_yolcunun_bilgisi)

```

```python
# Mühendislikte sütunların kaçıncı sırada olduğunu ezberlemek zordur. Bu yüzden loc kullanarak sütun isimleriyle de erişebiliriz.
yolcu_isim_yas = titanic_df.loc[0, ["Name", "Age"]]
print(yolcu_isim_yas)
```

```python
# Diyelim ki lk 10 yolcunun (0'dan 9'a kadar olan satırlar) sadece "Name" ve "Ticket" (Bilet numarası) bilgilerine ihtiyacımız var.

ilk_10_yolcu_bilgi = titanic_df.loc[0:9, ["Name", "Ticket"]]
print(ilk_10_yolcu_bilgi)
```
