Pandas ile kolayca gruplama yapabiliriz. Bunun için groupby kullanırız.

Aynı zamanda pandas ile işlem yaptığımız verilerin görselleştirmesi de önemlidir bunun için matplotlib kullanıcaz.

```python
import pandas as pd

link = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

titanic_df = pd.read_csv(link)
```

```python
# Genel şablonumuz şu şekildedir:
# DataFrame.groupby("Gruplama Sütunu")["İşlem Yapılacak Sütun"].islem()


siniflara_gore_fiyat = titanic_df.groupby("Pclass")["Fare"].mean()
print(siniflara_gore_fiyat)
```

ÖDEV:
Titanik efsanesinin o en bilindik tartışmasını veriyle kanıtlayalım: "Gerçekten de kadınlar erkeklere göre daha mı çok kurtarıldı?"

Yolcuları cinsiyetlerine göre gruplayıp, her grubun hayatta kalma oranının ortalamasını (mean) hesaplayan kodu nasıl yazarsın?

```python
cinsiyet_hayatta_kalan_ortalamasi = titanic_df.groupby("Sex")["Survived"].mean()
print(cinsiyet_hayatta_kalan_ortalamasi)
```

Kadınların yaklaşık %74'ü hayatta kalmış.
Erkeklerin yaklaşık %19'u hayatta kalmış.

Şimdi bunu grafiğe dönüştürelim

```python
import matplotlib.pyplot as plt

cinsiyet_hayatta_kalan_ortalamasi.plot(kind="bar", color=["pink", "blue"], title="Cinsiyete Göre Hayatta Kalma Oranı")
plt.xlabel("Cinsiyet")
plt.ylabel("Hayatta Kalma Oranı")
plt.show()
```
