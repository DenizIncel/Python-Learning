## Pandas Merge / Join / Concat

```python
import pandas as pd

# 2 Data frame oluşturalım.

students = pd.DataFrame({
    "student_id": [1, 2, 3, 4],
    "name": ["Ahmet", "Ayşe", "Mehmet", "Zeynep"]
})

scores = pd.DataFrame({
    "student_id": [1, 2, 4, 5],
    "score": [90, 85, 95, 80]
})

print("students:")
print(students)
print("\nscore:")
print(scores)
```

`merge` ile ikisini birleştirelim. Burada **student_id** bizim için ortak key.

**inner**: Sadece her 2 tabloda da ortak olan satırları getirir.

**outer**: Her 2 tablodaki tüm satırları getirir.

```python
inner_merge = pd.merge(students, scores, on="student_id", how="inner")
# on: hangi sütunları baz alanarak eşleşme yapacağını belirtiriz.
# how: birleşme türünü belirler. varsayılan inner'dir.
print(inner_merge)

outer_merge = pd.merge(students, scores, on="student_id", how="outer")
print("\n", outer_merge)
```

`join`: Genellikle indeks bazlı birleştirme için kullanılır, ama on parametresiyle sütun bazlı da yapılabilir.

`concat`: DataFrame'leri üst üste veya yan yana ekler. Verileri birleştirmek yerine satır/sütun eklemek için uygundur.

```python
students = pd.DataFrame({
    "student_id": [1, 2, 3, 4],
    "name": ["Ahmet", "Ayşe", "Mehmet", "Zeynep"]
})

new_students = pd.DataFrame({
    "student_id": [5,6],
    "name": ["Arda", "Kenan"],
    })


satir_ekleme = pd.concat([students, new_students]).set_index("student_id")
print(satir_ekleme)


# Yeni öğrenciler, diğer öğrencilerin altına satır olarak eklenir.
```

## Bazı onemli Pandas fonksiyonları

- `apply`: bir satırda veya sütunda fonksiyon çalıştırmak için kullanılır.
- `astype`: sütun tipini değiştirmek için kullanılır.
- `pivot_table`: veriyi satır/sütun bazında özetler.
- `fillna`: Eksik değerlere değer atar.

```python
sales = pd.DataFrame({
    "date": ["01.01.2025", "01.01.2025", "02.01.2025", "01.01.2025", "03.01.2025"], #tarihler
    "city": ["Ankara", "İstanbul", "Ankara", "İzmir", "İstanbul"], #şehirler
    "product": ["A", "B", "A", "B", "A"], #ürünler
    "quantity": [10, 5, 8, None, 12], #miktarlar
    "price": [20.5, 30.0, 20.5, 30.0, None] #fiyatlar
})

print("Orijinal sales DataFrame:")
print(sales)

# astype: quantity ve price sütunlarını doğru tipe çevirme
sales["quantity"] = sales["quantity"].astype("float")
sales["price"] = sales["price"].astype("float")

# fillna: eksik değerlere değer atama
sales["quantity"] = sales["quantity"].fillna(0)
sales["price"] = sales["price"].fillna(sales["price"].mean())

# apply: her satır için total hesaplama
sales["total"] = sales.apply(lambda row: row["quantity"] * row["price"], axis=1) 

# sort_values: toplam gelire göre sırala
sales_sorted = sales.sort_values(by="total", ascending=False)

print("\nTemizlenmiş ve total eklenmiş sales:")
print(sales_sorted)
```
