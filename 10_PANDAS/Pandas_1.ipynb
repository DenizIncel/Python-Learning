Pandas, Python'da veri analizi ve manipülasyonu için kullanılan güçlü bir kütüphanedir. Veri yapıları ve fonksiyonları sayesinde, büyük veri setlerini kolayca işleyebilir ve analiz edebilirsiniz. Pandas, özellikle DataFrame adlı veri yapısı ile bilinir, bu yapı sayesinde verileri tablo formatında düzenleyebilir ve üzerinde çeşitli işlemler gerçekleştirebilirsiniz.

```python
import pandas as pd
```

```python
# Önce bir sözlük oluşturalım.

veriler = {
    "Film Adı" : ["Inception", "Interstellar", "The Dark Knight", "Memento"],
    "Tür" : ["Bilim Kurgu", "Bilim Kurgu", "Aksiyon", "Gerilim"],
    "Yil" : [2010, 2014, 2008, 2000]
}

# Şimdi bu sözlüğü bir DataFrame'e dönüştürelim.

df1  = pd.DataFrame(veriler)

print(df1)
```

```python
df1 = df1.set_index("Film Adı")  # Film Adı sütununu index yapıyoruz. Önceden 0,1,2.. olan indexler film adları oldu.
print(df1)

# DataFrame'i bir CSV dosyasına kaydedelim.
df1.to_csv("filmler.csv")

# CSV dosyasını okuyalım ve bazı değişiklikler yapalım.
df2 = pd.read_csv("filmler.csv")
df2 = df2.set_index("Yil")
print(df2)
```

```python
# Pandas ile serilerde oluşturabiliriz. Bunlar DataFrame aksine tek bir sütun içerirler.

serie1 = pd.Series([10, 20, 30, 40])
print(serie1)

# index parametresi ile istediğimiz indexleri verebiliriz.
serie2 = pd.Series([10, 20, 30, 40] , index=["a", "b", "c", "d"])
print(serie2)
```
