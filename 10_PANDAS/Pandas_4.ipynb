Pandas ile sıralama gibi pek çok işlem yapabiliriz.

```python
import pandas as pd

# sort_values: Sıralama yapmak için kullanırız. Belirli bir sütuna göre sıralama yapar.
# ascending: Sıralamanın artan mı yoksa azalan mı olacağını belirler. True ise artan, False ise azalan sıralama yapar.

yaslar = [25, 30, 22, 28, 35]
yaslar = pd.Series(yaslar)
yaslar = yaslar.sort_values(ascending=False) # Yaşları küçükten büyüğe doğru sıralar.
print(yaslar)
```

Pandas ile bazı işlemler de yapabiliriz. Örneğin bilet ve kişi sayısını çarparak toplam fiyatı bulmak gibi.

```python
data = {
    "Ad": ["Tony", "Steve", "Bruce", "Natasha"],
    "Bilet Fiyatı": [15, 8.50, 7.00, 8.00],
    "Kişi Sayısı": [3, 2, 1, 1]
}

df = pd.DataFrame(data)
df = df.set_index("Ad")
print(df)

df = df.sort_values(by="Bilet Fiyatı", ascending=False) # by: Sıralama yapılacak sütunu belirtir.
print(df)

df["Toplam Fiyat"] = df["Bilet Fiyatı"] * df["Kişi Sayısı"]
print(df)
```

ÖDEV : Titanik tablomuzu düşünelim. Eğer bileti alan kişi (1) + eş/kardeş sayısı + ebeveyn/çocuk sayısı bize o biletle seyahat eden toplam kişi sayısını veriyorsa; kişi başına düşen gerçek bilet fiyatını bulup bunu Kisi_Basina_Fiyat adında yeni bir sütuna kaydetmek için nasıl bir Pandas kodu yazardın?

Bir yolcunun gemideki eş/kardeş sayısı SibSp sütununda, ebeveyn/çocuk sayısı ise Parch sütununda yer alıyor.

```python
import pandas as pd

link = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

titanic_df = pd.read_csv(link)

titanic_df["Kisi_Basina_Fiyat"] = titanic_df["Fare"] / ((titanic_df["SibSp"] + titanic_df["Parch"]) + 1)

titanic_df = (titanic_df[["Fare", "SibSp", "Parch", "Kisi_Basina_Fiyat"]])

titanic_df = titanic_df.sort_values(by="Kisi_Basina_Fiyat", ascending=False)

print(titanic_df.head(20))
```
