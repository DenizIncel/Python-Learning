Pandas ile dışarıdan devasa bir dosya okuyup bu verinin röntgenini çekelim. Pandas, bilgisayarındaki bir dosyayı okuyabildiği gibi, internetteki bir CSV dosyasını da indirmeden doğrudan okuyabilir.

Dünyanın en meşhur veri setlerinden biri olan Titanic yolcu listesini bilgisayarımıza çekip inceleyelim.

```python
import pandas as pd

# İnternetteki raw CSV dosyasının linki:
link = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

# read_csv komutu ile dosyayı okuyup bir DataFrame'e dönüştürüyoruz
titanic_df = pd.read_csv(link)

print(titanic_df.head(7))  # İlk 7 satırı görüntüle
# Eğer head() içini boş bırakırsak, varsayılan olarak ilk 5 satırı gösterir.
```

```python
print(titanic_df.info())  # DataFrame hakkında genel bilgi verir
# Non-null count: Her sütunda kaç tane geçerli (null olmayan) veri olduğunu gösterir.
# Bu DataFrame'de bazı sütunlarda eksikler var. En çok eksik veriye sahip olan Cabin sütunu gibi görünüyor.
```

Kurtarılamayacak kadar eksik bilgi olduğu için Cabin sütununu silelim.

Pandas'ta silme işlemi için .drop() komutunu kullanırız.Fakat Pandas'ın hedefimizin satır mı yoksa sütun mu olduğunu anlaması için bir "eksen" (axis) belirtmemiz gerekir:

**axis=0 -> Satır siler.**

**axis=1 -> Sütun siler.**

```python
# Cabin sütununu DataFrame'den kaldırıyoruz çünkü çok fazla eksik veri içeriyor.
titanic_df = titanic_df.drop(['Cabin'], axis=1) 
print(titanic_df.info())
```

Yaş, Titanik'te kimin kurtulacağını tahmin etmek için çok kritik bir bilgi ("önce kadınlar ve çocuklar" kuralı). Bu yüzden bu sütunu silip atamayız. O 177 kişiyi komple tablodan silmek de büyük bir veri kaybı olur.

İşte burada ikinci temizlik silahımız devreye giriyor: **Doldurma (fillna)**.

Yapay zeka modellerini eğitirken en yaygın taktik, bu tarz sayısal eksiklikleri veri setinin ortalamasıyla doldurmaktır. Böylece veri setinin genel yapısını bozmamış oluruz.

```python
# İlk adımımız gemideki mevcut yolcuların yaş ortalamasını almak olsun. Bunun için mean() fonksiyonunu kullanacağız.

ortalama_yas = titanic_df["Age"].mean()
print("Yaş ortalaması:", ortalama_yas)

# Age sütunundaki boşlukları yaş ortalaması ile dolduralım. Bunun için fillna() fonksiyonunu kullanacağız.

titanic_df["Age"] = titanic_df["Age"].fillna(ortalama_yas)

print(titanic_df.info())  # Artık Age sütununda eksik veri kalmadı. non-null count: 891 oldu.
```

**Pandas'ta info() tablonun teknik yapısını (boşluklar, veri tipleri) gösterirken, describe() komutu tablonun istatistiksel özetini çıkarır.**

describe() komutu sadece sayısal sütunları (yaş, fiyat vb.) alır ve saniyeler içinde ortalama, standart sapma, minimum ve maksimum gibi değerleri hesaplar. Özellikle yapay zeka ve mühendislik projelerinde verinin nasıl dağıldığını (örneğin sensörden gelen saçma veya aykırı değerler olup olmadığını) anlamak için ilk bakılan yerdir.

```python
print(titanic_df.describe())

#count sütununda 891 yolcu olduğunu görüyoruz. Bu, gemideki toplam yolcu sayısını gösterir.
#mean sütununda  ortalamasını görüyoruz. Örneğin ortalama yaş yaklaşık 29.7'dir.
#std sütununda standart sapmayı görüyoruz. Bu, yaşların ortalamadan ne kadar saptığını gösterir. Yaklaşık 14.5'tir.
#min ve max sütunlarında yaşların minimum ve maksimum değerlerini görüyoruz.
```
