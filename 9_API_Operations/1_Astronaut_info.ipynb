```python
import requests
import json
```

Python Requests, Python ile HTTP istekleri (GET, POST, PUT, DELETE vb.) göndermeyi son derece basit ve "insan odaklı" hale getiren popüler bir kütüphanedir.

```python
cevap = requests.get("http://api.open-notify.org/astros.json") #adresden bilgi aldık.

print(cevap.status_code) #cevabın durum kodunu yazdırdık.

veriler = cevap.json() #cevabın içeriğini json olarak aldık.

print(veriler)
```

```python
duzenli_veriler = json.dumps(veriler, indent=4)  #json verilerini  okunabilir hale getirmek için düzenledik.
print(duzenli_veriler)
```

```python
people_list = veriler["people"] #sadece astronotların bilgisini aldık.
print(people_list)
```

Bu karmaşık yapı yerine sadece astronotların isimlerini alt alta yazdırmak istiyorsak bu döngüyü kullanabiliriz:

```python
for i in veriler["people"]:
    print(i["name"]) #astronotların isimlerini yazdırdık.
```
